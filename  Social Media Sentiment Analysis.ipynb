```python
# Setup and Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from colorama import Fore, init
import plotly.express as px
```

```python
df = pd.read_csv("synthetic_social_media_data.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
## checking null values
np.sum(df.isnull().any(axis =1))
```

```python
## Count plot of sentiment labels

plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment Label', data=df, palette='coolwarm')
plt.title('Distribution of Sentiment Labels')
plt.show()
```

```python
## Number of Likes vs Sentiment Labels

plt.figure(figsize=(10, 6))
sns.boxplot(x='Sentiment Label', y='Number of Likes', data=df, palette='coolwarm')
plt.title('Number of Likes by Sentiment Label')
plt.show()
```

```python
## Relationship between Number of Likes, Shares, and Comments

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.pairplot(df[['Number of Likes', 'Number of Shares', 'Number of Comments']], diag_kind='kde', palette='coolwarm')
plt.show()
```

```python
## Likes vs Follower Count for Different Post Types

plt.figure(figsize=(10, 6))
sns.scatterplot(x='User Follower Count', y='Number of Likes', hue='Post Type', data=df, palette='coolwarm', alpha=0.6)
plt.title('Number of Likes vs Follower Count by Post Type')
plt.show()
```

```python
## Post Type Distribution by Sentiment Label

plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment Label', hue='Post Type', data=df, palette='coolwarm')
plt.title('Distribution of Post Types by Sentiment Label')
plt.show()
```

```python
## Number of Shares vs Sentiment

plt.figure(figsize=(10, 6))
sns.violinplot(x='Sentiment Label', y='Number of Shares', data=df, palette='Set2')
plt.title('Distribution of Number of Shares by Sentiment Label')
plt.show()
```

```python
## Heatmap of Correlations between Numeric Variables

plt.figure(figsize=(10, 6))
corr = df[['Number of Likes', 'Number of Shares', 'Number of Comments', 'User Follower Count']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Engagement Metrics')
plt.show()
```
